## Crear DATASET

Para comenzar el proyecto, creé una colección de imágenes que incluye páginas con y sin escritura, usando diferentes colores de fondo. Esto fue necesario ya que no contábamos con un conjunto de datos ya clasificado.
Para analizar cada imagen, seguí estos pasos:

1. Convertí las imágenes a escala de grises
2. Calculé el promedio de intensidad
3. Medí la variación de intensidad entre píxeles
4. Determiné qué porcentaje de la imagen tenía tonos oscuros

Finalmente, organicé toda esta información en un dataset, añadiendo una columna que indica si la página contiene texto (marcada como 1) o está en blanco (marcada como 0) y lo exporte como archivo csv.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os

def extract_features(image):
    """
    Extrae características básicas y relevantes de la imagen.
    """
    img_array = np.array(image)
    
    avg_intensity = np.mean(img_array)
    std_intensity = np.std(img_array)
    black_percentage = np.sum(img_array < 50) / img_array.size  # Consideramos píxeles con intensidad < 50 como "negros"
    
    return [avg_intensity, std_intensity, black_percentage]

def create_dataset(directory):
    data = []
    for filename in os.listdir(directory):
        if filename.endswith('.png'):
            img = Image.open(os.path.join(directory, filename)).convert('L')
            features = extract_features(img)
            is_written = 1 if 'written' in filename else 0
            data.append(features + [is_written])
    
    columns = ['avg_intensity', 'std_intensity', 'black_percentage', 'is_written']
    
    return pd.DataFrame(data, columns=columns)

# Crear el dataset
df = create_dataset('dataset/')

# Guardar el dataset en un archivo CSV
df.to_csv('dataset_imagenes_simple.csv', index=False)
```

## Entrenamiento

Para clasificar las imágenes, utilizaremos un perceptrón simple, que es un modelo de clasificación que aprendimos en clase. Este perceptrón trabajará con dos variables de entrada:

* El promedio de intensidad de la imagen
* La variación de intensidad entre píxeles

La elección de estas dos variables se basa en el análisis previo, donde demostraron ser efectivas para distinguir entre páginas escritas y no escritas.

Luego del entrenamiento se logró una correcta identificación de los datos ingresados. Además, podemos observar en la frontera de decisión cómo se consigue gráficamente la correcta clasificación entre las dos clases.

```python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    """
    Implementación de un perceptrón simple.

    Atributos:
        weights (ndarray): Pesos del perceptrón, incluyendo el bias.
        lr (float): Tasa de aprendizaje.
        epochs (int): Número de iteraciones para el entrenamiento.
    """
    
    def __init__(self, num_inputs: int, lr: float, epochs: int, pesos=None):
        """
        Inicializa el perceptrón.

        Args:
            num_inputs (int): Número de entradas.
            lr (float): Tasa de aprendizaje.
            epochs (int): Número de iteraciones para el entrenamiento.
            pesos (ndarray, optional): Pesos iniciales del perceptrón. 
                Si se proporciona, se utilizarán estos pesos.
        """
        if pesos is not None:
            self.weights = pesos  # Cargar pesos predefinidos
        else:
            self.weights = np.random.rand(num_inputs + 1)  # Peso extra para el bias
            
        self.lr = lr  # Tasa de aprendizaje
        self.epochs = epochs  # Número de iteraciones

    def act_fn(self, x: np.ndarray, funcion: str = 'step') -> np.ndarray:
        """Calcula la función de activación."""
        if funcion == 'step':    
            return np.where(x > 0, 1, 0)
        elif funcion == 'sigmoid':
            return 1 / (1 + np.exp(-x)) 
        elif funcion == 'relu':
            return np.where(x > 0, x, 0)
        elif funcion == 'tanh':
            return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
        else:
            raise ValueError("Función de activación no soportada.")

    def predict(self, inputs: np.ndarray) -> int:
        """Realiza una predicción basada en las entradas."""
        return self.act_fn(np.dot(inputs, self.weights[1:]) + self.weights[0])

    def compute_cost(self, in_train: np.ndarray, clases: np.ndarray) -> float:
        """Calcula el costo total (error cuadrático medio)."""
        predicciones = self.predict(in_train)
        error = clases - predicciones
        return np.mean(error ** 2)

    def train(self, in_train: np.ndarray, clases: np.ndarray, verbose: bool = False):
        """Entrena el perceptrón con las entradas y las clases."""
        errores = []
        for _ in range(self.epochs):
            e = 0
            for entrada, clase in zip(in_train, clases):
                prediccion = self.predict(entrada)
                error = clase - prediccion
                self.weights[0] += self.lr * error
                self.weights[1:] += self.lr * error * entrada
                
                e += np.abs(error)
            errores.append(e)
            
            # Calcular y almacenar el costo al final de cada época
            costo = self.compute_cost(in_train, clases)
            print(f"Época {_ + 1}: Costo = {costo}")

        if verbose:
            plt.figure()
            plt.plot(errores)
            plt.title('Errores durante el entrenamiento')
            plt.xlabel('Epochs')
            plt.ylabel('Error')
            plt.show()

    def get_weights(self) -> np.ndarray:
        """Recupera los pesos del perceptrón."""
        return self.weights

    def plot_decision_boundary(self, in_train: np.ndarray, clases: np.ndarray):
        """Plotea las entradas y la frontera de decisión del perceptrón."""
        # Definir el rango para el gráfico
        x_min, x_max = in_train[:, 0].min() - 1, in_train[:, 0].max() + 1
        y_min, y_max = in_train[:, 1].min() - 1, in_train[:, 1].max() + 1
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), 
                             np.linspace(y_min, y_max, 100))

        # Realizar predicciones para cada punto del meshgrid
        Z = self.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)

        # Crear el gráfico
        plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
        plt.scatter(in_train[:, 0], in_train[:, 1], c=clases, edgecolors='k', marker='o', cmap=plt.cm.coolwarm)
        plt.xlabel('Entrada 1')
        plt.ylabel('Entrada 2')
        plt.title('Frontera de decisión del Perceptrón')
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('dataset_imagenes_simple.csv')


x = df[['avg_intensity', 'std_intensity']].to_numpy()
y = df['is_written'].to_numpy()

ia = Perceptron(num_inputs=2, lr=1, epochs=2)
ia.train(x, y, True)
for entradas, label in zip(x, y):
  prediccion = ia.predict(entradas)
  print('Entradas:', entradas, ' Correcto:', label, ' Predicción:', prediccion)
ia.plot_decision_boundary(x,y)
```

## Pruebas

Luego de obtener el modelo, realizo pruebas con imagenes reales, se logran resultados buenos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os

def mostrar_clasificacion(ruta_imagen, modelo):
    # Cargar y procesar la imagen
    img = Image.open(ruta_imagen).convert('L')
    img_array = np.array(img)
    
    # Calcular características
    avg_intensity = np.mean(img_array)
    std_intensity = np.std(img_array)
    
    # Realizar predicción
    x = np.array([[avg_intensity, std_intensity]])
    prediccion = modelo.predict(x)
    
    # Configurar la visualización
    plt.figure(figsize=(12, 5))
    
    # Mostrar imagen original
    imgOriginal = Image.open(ruta_imagen)

    plt.subplot(1, 2, 1)
    plt.imshow(imgOriginal, cmap='gray')
    plt.title('Imagen Original')
    plt.axis('off')
    
    # Mostrar resultado
    plt.subplot(1, 2, 2)
    plt.text(0.5, 0.5, 
             f'Resultado:\n{"ESTA ESCRITA O DIBUJADA" if prediccion[0] == 1 else "NO ESTA ESCRITA O DIBUJADA"}\n\n'
             f'Intensidad promedio: {avg_intensity:.2f}\n'
             f'Desviacion: {std_intensity:.2f}',
             horizontalalignment='center',
             verticalalignment='center',
             fontsize=14)
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Pruebas
mostrar_clasificacion('imagen1.jpeg', ia)
mostrar_clasificacion('imagen2.jpeg', ia)
mostrar_clasificacion('imagen3.jpeg', ia)
mostrar_clasificacion('imagen4.png', ia)
```
